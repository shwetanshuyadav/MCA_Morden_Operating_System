```python
def first_fit(blocks, processes):
    allocation = [-1] * len(processes)
    for i in range(len(processes)):
        for j in range(len(blocks)):
            if blocks[j] >= processes[i]:
                allocation[i] = j
                blocks[j] -= processes[i]
                break
    return allocation

def best_fit(blocks, processes):
    allocation = [-1] * len(processes)
    for i in range(len(processes)):
        best_fit_index = -1
        for j in range(len(blocks)):
            if blocks[j] >= processes[i]:
                if best_fit_index == -1 or blocks[j] < blocks[best_fit_index]:
                    best_fit_index = j
        if best_fit_index != -1:
            allocation[i] = best_fit_index
            blocks[best_fit_index] -= processes[i]
    return allocation

def worst_fit(blocks, processes):
    allocation = [-1] * len(processes)
    for i in range(len(processes)):
        worst_fit_index = -1
        for j in range(len(blocks)):
            if blocks[j] >= processes[i]:
                if worst_fit_index == -1 or blocks[j] > blocks[worst_fit_index]:
                    worst_fit_index = j
        if worst_fit_index != -1:
            allocation[i] = worst_fit_index
            blocks[worst_fit_index] -= processes[i]
    return allocation
# Define the memory blocks and processes
blocks = [100, 500, 200, 300, 600]
processes = [212, 417, 112, 426]

# Run the algorithms and print the results
print("First Fit:", first_fit(blocks.copy(), processes))
print("Best Fit:", best_fit(blocks.copy(), processes))
print("Worst Fit:", worst_fit(blocks.copy(), processes))
```
