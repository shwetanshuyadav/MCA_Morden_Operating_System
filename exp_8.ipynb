```python
class PriorityProcess:
    def __init__(self, pid, burst_time, priority):
        self.pid = pid
        self.burst_time = burst_time
        self.priority = priority
        self.waiting_time = 0
        self.turnaround_time = 0


def priority_scheduling(processes):
    processes.sort(key=lambda x: x.priority)

    total_time = 0
    for i, process in enumerate(processes):
        if i == 0:
            process.waiting_time = 0
        else:
            process.waiting_time = total_time

        total_time += process.burst_time
        process.turnaround_time = process.waiting_time + process.burst_time

    return processes


processes = [
    PriorityProcess(1, 10, 3),
    PriorityProcess(2, 1, 1),
    PriorityProcess(3, 2, 4),
    PriorityProcess(4, 1, 5),
    PriorityProcess(5, 5, 2),
]

scheduled_processes = priority_scheduling(processes)

for p in scheduled_processes:
    print(
        f"Process {p.pid}: "
        f"Priority={p.priority}, "
        f"Burst Time={p.burst_time}, "
        f"Waiting Time={p.waiting_time}, "
        f"Turnaround Time={p.turnaround_time}"
    )

waiting_times = [p.waiting_time for p in scheduled_processes]
turnaround_times = [p.turnaround_time for p in scheduled_processes]
```
