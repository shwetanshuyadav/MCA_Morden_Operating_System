```python
def fcfs_scheduling(processes):
    n = len(processes)
    processes = sorted(processes, key=lambda x: x['arrival_time'])

    start_time = []
    completion_time = []
    waiting_time = []
    turnaround_time = []

    current_time = 0

    for i in range(n):
        if current_time < processes[i]['arrival_time']:
            current_time = processes[i]['arrival_time']

        start_time.append(current_time)
        current_time += processes[i]['burst_time']
        completion_time.append(current_time)
        turnaround_time.append(
            completion_time[i] - processes[i]['arrival_time']
        )
        waiting_time.append(
            turnaround_time[i] - processes[i]['burst_time']
        )

    print("Process\tArrival\tBurst\tStart\tCompletion\tWaiting\tTurnaround")
    for i in range(n):
        print(
            f"{processes[i]['pid']}\t{processes[i]['arrival_time']}\t"
            f"{processes[i]['burst_time']}\t{start_time[i]}\t"
            f"{completion_time[i]}\t{waiting_time[i]}\t"
            f"{turnaround_time[i]}"
        )


process_list = [
    {'pid': 'P1', 'arrival_time': 0, 'burst_time': 4},
    {'pid': 'P2', 'arrival_time': 1, 'burst_time': 3},
    {'pid': 'P3', 'arrival_time': 2, 'burst_time': 1},
    {'pid': 'P4', 'arrival_time': 3, 'burst_time': 2},
]

fcfs_scheduling(process_list)
```
