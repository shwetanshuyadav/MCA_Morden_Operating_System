```python
# Implementation of FCFS and SSTF Disk Scheduling Algorithms

def fcfs(requests, head):
    movement = 0
    current = head

    for r in requests:
        movement += abs(r - current)
        current = r

    return movement
def sstf(requests, head):
    movement = 0
    current = head
    req = requests.copy()

    while req:
        closest = min(req, key=lambda x: abs(x - current))
        movement += abs(closest - current)
        current = closest
        req.remove(closest)

    return movement

requests = [98, 183, 37, 122, 14, 124, 65, 67]
head = 53

print("FCFS Total Head Movement =", fcfs(requests, head))
print("SSTF Total Head Movement =", sstf(requests, head))
```
