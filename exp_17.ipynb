```python
class ResourceManager:
    def __init__(self, resources):
        self.resources = sorted(resources)
        self.allocated = {}

    def request(self, process, req_resources):
        req_resources = sorted(req_resources)

        # Check if resources exist
        for r in req_resources:
            if r not in self.resources:
                print(f"Resource {r} does not exist")
                return False

        # Check if resource is already allocated
        for r in req_resources:
            if r in self.allocated.values():
                print(f"Process {process} waiting for resource {r}")
                return False

        # Allocate resources
        for r in req_resources:
            self.allocated[process] = r
            print(f"Process {process} allocated resource {r}")

        return True

    def release(self, process):
        if process in self.allocated:
            print(f"Process {process} released resource {self.allocated[process]}")
            del self.allocated[process]
        else:
            print(f"Process {process} has no resources to release")


# Driver code
rm = ResourceManager([1, 2, 3])

rm.request("P1", [1, 2])
rm.request("P2", [2, 3])
rm.release("P1")
rm.request("P2", [2, 3])
```
