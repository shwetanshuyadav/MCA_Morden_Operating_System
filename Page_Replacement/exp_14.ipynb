```python
# File access pattern analysis to suggest cache strategy
def suggest_cache_strategy(accesses):
    if len(accesses) < 2:
        return "LRU Cache"

    sequential = 0
    random = 0

    for i in range(1, len(accesses)):
        if accesses[i] == accesses[i - 1] + 1:
            sequential += 1
        else:
            random += 1

    if sequential > random:
        return "Read-Ahead / FIFO Cache"
    else:
        return "LRU Cache"


file_accesses = [10, 11, 12, 13, 14, 20, 5, 30, 6, 7]

strategy = suggest_cache_strategy(file_accesses)

print("File Access Pattern:", file_accesses)
print("Suggested Cache Strategy:", strategy)
```
