```python
def predictive_lru(pages, frames):
    memory = []
    faults = 0
    history = {} 
    for page in pages:
        history[page] = history.get(page, 0) + 1
        if page not in memory:
            faults += 1
            if len(memory) < frames:
                memory.append(page)
            else:
                victim = memory[0]
                min_freq = history.get(victim, 0)
                for m in memory:
                    current_freq = history.get(m, 0)
                    if current_freq < min_freq:
                        min_freq = current_freq
                        victim = m
                memory.remove(victim)
                memory.append(page)
        else:
            memory.remove(page)
            memory.append(page)
        print(f"Page: {page}, Frames: {memory}")
    print(f"Total Page Faults (Predictive LRU): {faults}")

# Example Usage:
pages_example = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3, 2, 1, 2, 0] 
frames_example = 3

predictive_lru(pages_example, frames_example)
```
