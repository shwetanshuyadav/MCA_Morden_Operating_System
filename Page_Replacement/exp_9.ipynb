```python
def fifo_page_replacement(pages, capacity):
    memory = []
    page_faults = 0

    for page in pages:
        if page not in memory:
            page_faults += 1
            if len(memory) < capacity:
                memory.append(page)
            else:
                memory.pop(0) # Remove oldest page
                memory.append(page)
        
        # Corrected print statement
        print(f"Page: {page} -> Memory: {memory}") 

    print("Total Page Faults (FIFO):", page_faults)

pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3]
capacity = 3

# Run the corrected function
fifo_page_replacement(pages, capacity)
```

```python
def lru_page_replacement(pages, capacity):
    memory = []
    page_faults = 0
    print(f"--- Simulating LRU with Capacity: {capacity} ---")
    for page in pages:
        if page not in memory:
            page_faults += 1
            if len(memory) < capacity:
                memory.append(page)
            else:
                memory.pop(0)
                memory.append(page)
        else:
            memory.remove(page)
            memory.append(page)
        print(f"Page reference: {page} -> Current Memory State: {memory}")
    print("Total Page References:", len(pages))
    print("Total Page Faults (LRU):", page_faults)
# Example usage:
pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3]
capacity = 3
lru_page_replacement(pages, capacity)
```
