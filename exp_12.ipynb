```python
def optimal(pages, frames):
    memory = []
    faults = 0

    for i in range(len(pages)):
        # Page fault
        if pages[i] not in memory:
            faults += 1

            if len(memory) < frames:
                memory.append(pages[i])
            else:
                future = []

                for m in memory:
                    if m in pages[i+1:]:
                        future.append(pages[i+1:].index(m))
                    else:
                        future.append(float('inf'))

                # Replace the page with the farthest future use
                memory[future.index(max(future))] = pages[i]

    return faults


pages = [7, 0, 1, 2, 0, 3, 0, 4, 2, 3, 0, 3]
frames = 3

print("Optimal Page Faults:", optimal(pages, frames))
```
