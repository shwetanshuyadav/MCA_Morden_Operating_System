```python
import psutil

def top_processes(n=5):
    processes = []
    for proc in psutil.process_iter(['pid', 'name', 'cpu_times']):
        try:
            info = proc.info
            total_cpu = sum(info['cpu_times']) if info['cpu_times'] else 0
            processes.append({'pid': info['pid'], 'name': info['name'], 'cpu': total_cpu})
        except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
            continue
        
    top = sorted(processes, key=lambda p: p['cpu'], reverse=True)[:n]
    print("{:<25} {:<10}".format("Process Name", "PID"))
    for proc in top:
        print("{:<25} {:<10}".format(proc['name'], proc['pid']))

if __name__ == "__main__":
    top_processes(5)
```
